```python

# -*- coding: utf-8 -*-
# Requisitos: pandas, openpyxl
# pip install pandas openpyxl

import os, re
from datetime import datetime
import pandas as pd

# === Ruta del archivo ===
# INPUT_PATH = r"C:\Users\TU_USUARIO\Downloads\TIGIE.xlsx"
INPUT_PATH = "raw_data/TIGIE.xlsx"

BASE_DIR = os.path.dirname(os.path.abspath(__file__))
INPUT_PATH = os.path.join(BASE_DIR, INPUT_PATH)

# ---------- utilidades ----------
def _norm(s: str) -> str:
    import unicodedata
    s = "" if s is None else str(s)
    s = "".join(c for c in unicodedata.normalize("NFD", s.lower())
                if unicodedata.category(c) != "Mn")
    return re.sub(r"\s+", " ", s).strip()

def _strip_weird_spaces(s: str) -> str:
    # elimina NBSP, NO-BREAK, thin spaces, etc.
    return s.replace("\u00A0", " ").replace("\u2007", " ").replace("\u202F", " ").strip()

def _extract_two_level(value: object, tolerant: bool = False) -> str | None:
    """
    Devuelve 'xx.xx' si el valor representa 2 niveles.
    - Acepta: '01.01', '0101', '1.1', '  01 . 01  '
    - Si tolerant=True: recorta cadenas con 3+ niveles a sus 2 primeros ('0101.29.02' -> '01.01')
    """
    if value is None:
        return None
    s = _strip_weird_spaces(str(value))

    # Mantener solo d√≠gitos y puntos (pero conservar el orden)
    s = "".join(ch for ch in s if ch.isdigit() or ch == ".")
    if not s:
        return None

    # Caso 4 d√≠gitos contiguos -> xx.xx
    if re.fullmatch(r"\d{4}", s):
        return f"{int(s[:2]):02d}.{int(s[2:]):02d}"

    parts = [p for p in s.split(".") if p != ""]
    # Si hay puntos:
    if parts:
        if len(parts) == 2 and all(1 <= len(p) <= 2 for p in parts):
            return f"{int(parts[0]):02d}.{int(parts[1]):02d}"
        if tolerant and len(parts) >= 2:  # recorta a los dos primeros niveles
            return f"{int(parts[0]):02d}.{int(parts[1]):02d}"
        return None

    # Si no hay puntos, pero hay de 1 a 3 d√≠gitos -> no v√°lido
    return None

# ---------- lectura + detecci√≥n de cabecera ----------
try:
    raw = pd.read_excel(INPUT_PATH, sheet_name=0, header=None, dtype=object)
except FileNotFoundError:
    raise SystemExit(f"‚ùå No se encontr√≥ el archivo: {INPUT_PATH}")
except Exception as e:
    raise SystemExit(f"‚ùå Error leyendo el Excel: {e}")

CAND_FRAC = ["fraccion arancelaria", "fraccion", "fa", "hs code", "hs", "partida", "tarifa"]
CAND_DESC = ["descripcion", "descripci√≥n", "description", "desc"]

header_row = None
for i in range(min(80, len(raw))):
    norms = [_norm(v) for v in raw.iloc[i].astype(str).tolist()]
    if any(any(k in n for k in CAND_FRAC) for n in norms) and \
       any(any(k in n for k in CAND_DESC) for n in norms):
        header_row = i
        break
if header_row is None:
    header_row = 4  # respaldo

header = raw.iloc[header_row].tolist()
df = raw.iloc[header_row + 1:].copy()
df.columns = header
df = df.reset_index(drop=True)

# ---------- localizar columnas ----------
norm_map = {_norm(c): c for c in df.columns}
col_frac = next((norm_map[n] for n in norm_map if any(k in n for k in CAND_FRAC)), None)
col_desc = next((norm_map[n] for n in norm_map if any(k in n for k in CAND_DESC)), None)
if not col_frac:
    raise SystemExit(f"‚ùå No encontr√© la columna de Fracci√≥n Arancelaria. Columnas: {list(df.columns)}")
if not col_desc:
    raise SystemExit(f"‚ùå No encontr√© la columna de Descripci√≥n. Columnas: {list(df.columns)}")

# Forzar a string la columna y limpiar espacios invisibles
df[col_frac] = df[col_frac].astype(str).map(_strip_weird_spaces)

# ---------- modo estricto ----------
tmp = df[[col_frac, col_desc]].copy()
tmp["__fa__"] = tmp[col_frac].map(lambda v: _extract_two_level(v, tolerant=False))
resultado = (
    tmp.loc[tmp["__fa__"].notna(), [col_frac, col_desc]]
       .rename(columns={col_frac: "Fracci√≥n Arancelaria", col_desc: "Descripci√≥n"})
       .drop_duplicates()
       .reset_index(drop=True)
)

# Si no encontr√≥ nada, intenta modo TOLERANTE recortando a 2 niveles
tolerant_used = False
if resultado.empty:
    tolerant_used = True
    tmp["__fa__"] = tmp[col_frac].map(lambda v: _extract_two_level(v, tolerant=True))
    resultado = (
        tmp.loc[tmp["__fa__"].notna(), [col_frac, col_desc]]
           .rename(columns={col_frac: "Fracci√≥n Arancelaria", col_desc: "Descripci√≥n"})
           .drop_duplicates()
           .reset_index(drop=True)
    )

# ---------- exportaci√≥n / mensajes ----------
out_dir = os.path.join(BASE_DIR, "output")
os.makedirs(out_dir, exist_ok=True)
fecha = datetime.now().strftime("%Y%m%d")
out_path = os.path.join(out_dir, f"fracciones_4digitos_{fecha}.xlsx")

if resultado.empty:
    # Muestra ejemplos para depurar
    sample_raw = (df[col_frac].astype(str).head(15)).tolist()
    sample_clean = [ _extract_two_level(x, tolerant=True) for x in sample_raw ]
    print("‚ö†Ô∏è No se pudieron detectar fracciones de 2 niveles ni en modo tolerante.")
    print("üîé Muestra de valores crudos:", sample_raw)
    print("üîß Normalizados (tolerante):", sample_clean)
else:
    try:
        resultado.to_excel(out_path, index=False)
    except Exception as e:
        raise SystemExit(f"‚ùå Error exportando el Excel: {e}")

    if tolerant_used:
        print("‚ÑπÔ∏è Se activ√≥ el MODO TOLERANTE: se recortaron c√≥digos con 3+ niveles a los primeros 2 (xx.xx).")
    print(f"‚úÖ Listo. Registros exportados: {len(resultado)}")
    print(f"üìÑ Archivo generado: {out_path}")
```
