```python
import streamlit as st
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

st.set_page_config(page_title="Diferencia Fecha SO KMEX vs Fecha de PO", page_icon=":bar_chart:", layout="centered")

st.header("An치lisis: Fecha SO KMEX vs Fecha de PO (d칤as naturales)")
st.write("Sube tu archivo Excel")

archivo = st.file_uploader("Sube tu archivo Excel", type=["xlsx"])

if archivo:
    df = pd.read_excel(archivo)

        # Convertir columnas a fecha
            df['Fecha SO KMEX'] = pd.to_datetime(df['Fecha SO KMEX'], errors='coerce')
                df['Fecha de PO'] = pd.to_datetime(df['Fecha de PO'], errors='coerce')

                    # Filtrar filas v치lidas
                        df_so_po = df[df['Fecha SO KMEX'].notna() & df['Fecha de PO'].notna()].copy()

                            # Calcular diferencia en d칤as naturales (Fecha de PO - Fecha SO KMEX)
                                df_so_po['Diferencia d칤as naturales'] = (df_so_po['Fecha de PO'] - df_so_po['Fecha SO KMEX']).dt.days

                                    # Sem치foro: On time <= 3 d칤as naturales
                                        def status_so_po(dias):
                                                if pd.isna(dias):
                                                            return 'Sin fecha'
                                                                    elif dias <= 3:
                                                                                return 'On time'
                                                                                        elif dias > 3:
                                                                                                    return 'Delay'
                                                                                                            else:
                                                                                                                        return 'Revisar'
                                                                                                                            df_so_po['Sem치foro SO_PO'] = df_so_po['Diferencia d칤as naturales'].apply(status_so_po)

                                                                                                                                num_so_po = len(df_so_po)

                                                                                                                                    # Mostrar tabla SO vs PO
                                                                                                                                        st.subheader("Tabla (Fecha SO KMEX vs Fecha de PO)")
                                                                                                                                            st.dataframe(df_so_po[['Fecha SO KMEX', 'Fecha de PO', 'Diferencia d칤as naturales', 'Sem치foro SO_PO']])

                                                                                                                                                # Gauge chart
                                                                                                                                                    st.subheader("Diferencia Fecha de PO vs Fecha SO KMEX (d칤as naturales)")

                                                                                                                                                        if not df_so_po['Diferencia d칤as naturales'].dropna().empty:
                                                                                                                                                                promedio_so_po = df_so_po['Diferencia d칤as naturales'].mean()
                                                                                                                                                                    else:
                                                                                                                                                                            promedio_so_po = 0

                                                                                                                                                                                min_so_po = min(0, int(df_so_po['Diferencia d칤as naturales'].min()))
                                                                                                                                                                                    max_so_po = max(6, int((df_so_po['Diferencia d칤as naturales'].max() // 2 + 1) * 2))
                                                                                                                                                                                        tick_vals_so_po = list(range(min_so_po, max_so_po+1, 1))

                                                                                                                                                                                            fig_so_po = go.Figure(go.Indicator(
                                                                                                                                                                                                    mode = "gauge+number",
                                                                                                                                                                                                            value = promedio_so_po,
                                                                                                                                                                                                                    number = {'suffix': " d칤as", 'font': {'size': 36}},
                                                                                                                                                                                                                            title = {'text': "Fecha de PO vs Fecha SO KMEX"},
                                                                                                                                                                                                                                    gauge = {
                                                                                                                                                                                                                                                'axis': {'range': [min_so_po, max_so_po], 'tickmode': 'array', 'tickvals': tick_vals_so_po, 'tickfont': {'size': 16}},
                                                                                                                                                                                                                                                            'steps': [
                                                                                                                                                                                                                                                                            {'range': [min_so_po, 3], 'color': "#ABEBC6"},    # Verde: On Time
                                                                                                                                                                                                                                                                                            {'range': [3, max_so_po], 'color': "#F5B7B1"}     # Rojo: Delay
                                                                                                                                                                                                                                                                                                        ],
                                                                                                                                                                                                                                                                                                                    'bar': {'color': "black", 'thickness': 0.25}
                                                                                                                                                                                                                                                                                                                            }
                                                                                                                                                                                                                                                                                                                                ))

                                                                                                                                                                                                                                                                                                                                    st.plotly_chart(fig_so_po)
                                                                                                                                                                                                                                                                                                                                        st.markdown(
                                                                                                                                                                                                                                                                                                                                                f"<div style='text-align:center; font-size:18px; margin-top:-30px; margin-bottom:10px;'><b>Promedio considerando {num_so_po} diferencias</b></div>",
                                                                                                                                                                                                                                                                                                                                                        unsafe_allow_html=True
                                                                                                                                                                                                                                                                                                                                                            )

                                                                                                                                                                                                                                                                                                                                                                # Percentiles como texto SO_PO
                                                                                                                                                                                                                                                                                                                                                                    serie_so_po = df_so_po['Diferencia d칤as naturales'].dropna()
                                                                                                                                                                                                                                                                                                                                                                        percentil_50_so_po = serie_so_po.quantile(0.5)
                                                                                                                                                                                                                                                                                                                                                                            percentil_75_so_po = serie_so_po.quantile(0.75)
                                                                                                                                                                                                                                                                                                                                                                                percentil_90_so_po = serie_so_po.quantile(0.9)
                                                                                                                                                                                                                                                                                                                                                                                    percentil_95_so_po = serie_so_po.quantile(0.95)

                                                                                                                                                                                                                                                                                                                                                                                        st.subheader("Resumen de percentiles SO vs PO")
                                                                                                                                                                                                                                                                                                                                                                                            st.write(f"Percentil 50 (Mediana): **{percentil_50_so_po:.2f}** d칤as")
                                                                                                                                                                                                                                                                                                                                                                                                st.write(f"Percentil 75: **{percentil_75_so_po:.2f}** d칤as")
                                                                                                                                                                                                                                                                                                                                                                                                    st.write(f"Percentil 90: **{percentil_90_so_po:.2f}** d칤as")
                                                                                                                                                                                                                                                                                                                                                                                                        st.write(f"Percentil 95: **{percentil_95_so_po:.2f}** d칤as")

                                                                                                                                                                                                                                                                                                                                                                                                            # Histograma SO_PO
                                                                                                                                                                                                                                                                                                                                                                                                                st.subheader("Histograma de diferencia de d칤as naturales PO vs SO KMEX")
                                                                                                                                                                                                                                                                                                                                                                                                                    fig_hist_so_po = px.histogram(df_so_po, x='Diferencia d칤as naturales',
                                                                                                                                                                                                                                                                                                                                                                                                                                                      nbins=10,
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        title="Histograma (frecuencia por valor)")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            st.plotly_chart(fig_hist_so_po)

                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                # Tendencia temporal (promedio por mes)
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    st.subheader("Tendencia hist칩rica PO vs SO KMEX (promedio de d칤as naturales por mes)")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        df_so_po['A침oMes'] = df_so_po['Fecha de PO'].dt.to_period('M').astype(str)
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            trend_so_po = df_so_po.groupby('A침oMes')['Diferencia d칤as naturales'].mean().reset_index()
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                fig_trend_so_po = px.line(trend_so_po, x='A침oMes', y='Diferencia d칤as naturales', markers=True,
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                              title="Promedio de d칤as naturales por mes")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                  st.plotly_chart(fig_trend_so_po)

                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      st.info("Dashboard generado por Kai 游뱄")

                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      else:
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          st.warning("Sube un archivo para ver el dashboard.")
```
