```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import config

st.logo(
    image=config.horizontal,
        size = "large",
            icon_image=config.icon
                )

                st.set_page_config(page_title="Sem치foro Aduana", page_icon=":bar_chart:", layout="centered")
                st.title("KPI - Logistic Process 1")

                if "df_base" in st.session_state:
                    df = st.session_state["df_base"]
                    else:
                        st.warning("Primero ve a 'Supply Chain Process - Order Management' para cargar el archivo.")

                        # Convertir a fechas
                        df['EXW Real'] = pd.to_datetime(df['EXW Real'], errors='coerce')
                        df['Fecha entrada Aduana'] = pd.to_datetime(df['Fecha entrada Aduana'], errors='coerce')

                        # Filtrar filas v치lidas (con ambas fechas)
                        df_validas = df[
                            df['EXW Real'].notna() &
                                df['Fecha entrada Aduana'].notna() &
                                    (df['Fecha entrada Aduana'] >= pd.to_datetime("1971-01-01"))
                                        ].copy()

                                        # Calcular diferencia en d칤as h치biles
                                        df_validas['Diferencia d칤as h치biles'] = df_validas.apply(
                                            lambda x: np.busday_count(
                                                    x['EXW Real'].date(),
                                                            x['Fecha entrada Aduana'].date()
                                                                    ) if pd.notna(x['EXW Real']) and pd.notna(x['Fecha entrada Aduana']) else np.nan,
                                                                        axis=1
                                                                            )

                                                                            # Sem치foro nuevo: On time <= 8 d칤as h치biles
                                                                            def status_sem치foro(dias):
                                                                                if pd.isna(dias):
                                                                                        return 'Sin fecha'
                                                                                            elif dias <= 8:
                                                                                                    return 'On time'
                                                                                                        elif dias > 8:
                                                                                                                return 'Delay'
                                                                                                                    else:
                                                                                                                            return 'Revisar'
                                                                                                                            df_validas['Sem치foro'] = df_validas['Diferencia d칤as h치biles'].apply(status_sem치foro)

                                                                                                                            num_filas = len(df_validas)

                                                                                                                            df_validas['mes'] = df_validas['Fecha de PO'].dt.to_period('M')

                                                                                                                            meses_opciones = ["Todos"] + [str(m) for m in sorted(df_validas['mes'].dropna().unique())]
                                                                                                                            mes_elegido = st.selectbox("游늰 Selecciona un mes:", options=meses_opciones)

                                                                                                                            if mes_elegido != "Todos":
                                                                                                                                df_filtrado = df_validas[df_validas['mes'].astype(str) == mes_elegido]
                                                                                                                                else:
                                                                                                                                    df_filtrado = df_validas

                                                                                                                                    # Mostrar tabla
                                                                                                                                    st.subheader("EXW Real vs Fecha entrada Aduana")
                                                                                                                                    st.dataframe(df_filtrado[['PO', 'EXW Real', 'Fecha entrada Aduana', 'Diferencia d칤as h치biles', 'Sem치foro']])

                                                                                                                                    # Gauge chart (opcional)
                                                                                                                                    st.subheader("Sem치foro de EXW Real vs Fecha entrada Aduana (d칤as h치biles)")

                                                                                                                                    if not df_filtrado['Diferencia d칤as h치biles'].dropna().empty:
                                                                                                                                        promedio_dias = df_filtrado['Diferencia d칤as h치biles'].mean().round(0)
                                                                                                                                        else:
                                                                                                                                            promedio_dias = 0

                                                                                                                                            minimo = 0
                                                                                                                                            maximo = max(18, int((df_filtrado['Diferencia d칤as h치biles'].max() // 5 + 1) * 5))

                                                                                                                                            tick_vals = list(range(minimo, maximo+1, 2))

                                                                                                                                            fig = go.Figure(go.Indicator(
                                                                                                                                                mode = "gauge+number+delta",
                                                                                                                                                    delta={'reference': 8, 'increasing': {'color': "red"}, 'decreasing': {'color': "green"}},
                                                                                                                                                        value = promedio_dias,
                                                                                                                                                            number = {'suffix': " d칤as h치biles", 'font': {'size': 36}},
                                                                                                                                                                title = {'text': "EXW Real vs Fecha entrada Aduana"},
                                                                                                                                                                    gauge = {
                                                                                                                                                                            'axis': {'range': [minimo, maximo], 'tickmode': 'array', 'tickvals': tick_vals, 'tickfont': {'size': 16}},
                                                                                                                                                                                    'steps': [
                                                                                                                                                                                                {'range': [minimo, 8], 'color': 'green'},
                                                                                                                                                                                                            {'range': [8, 12], 'color': 'yellow'},
                                                                                                                                                                                                                        {'range': [12, 16], 'color': 'orange'},
                                                                                                                                                                                                                                    {'range': [16, maximo], 'color': 'red'}
                                                                                                                                                                                                                                            ],
                                                                                                                                                                                                                                                    'bar': {'color': "black", 'thickness': 0.25}
                                                                                                                                                                                                                                                            }
                                                                                                                                                                                                                                                                ))

                                                                                                                                                                                                                                                                st.plotly_chart(fig)
                                                                                                                                                                                                                                                                st.markdown(
                                                                                                                                                                                                                                                                    f"Promedio considerando {num_filas} diferencias(fechas confirmadas de entrada a aduana)",
                                                                                                                                                                                                                                                                        unsafe_allow_html=True
                                                                                                                                                                                                                                                                            )

                                                                                                                                                                                                                                                                            # ======= SCATTER PLOT DE PERCENTILES DE PROMEDIOS =======
                                                                                                                                                                                                                                                                            st.subheader("Curvas de percentiles mensuales (d칤as h치biles)")

                                                                                                                                                                                                                                                                            # Creamos columna de mes/a침o (por si acaso)
                                                                                                                                                                                                                                                                            df_validas['A침oMes'] = df_validas['Fecha de PO'].dt.to_period('M').astype(str)

                                                                                                                                                                                                                                                                            percentiles = [3, 15, 50, 85, 97]
                                                                                                                                                                                                                                                                            percentile_labels = ['P3', 'P15', 'P50', 'P85', 'P97']
                                                                                                                                                                                                                                                                            colors = ['#FFA07A', '#F4D03F', '#27AE60', '#F39C12', '#C0392B']

                                                                                                                                                                                                                                                                            # Calculamos percentiles mensuales
                                                                                                                                                                                                                                                                            df_percentiles = (
                                                                                                                                                                                                                                                                                df_validas.groupby('A침oMes')['Diferencia d칤as h치biles']
                                                                                                                                                                                                                                                                                    .apply(lambda x: pd.Series([x.quantile(p/100) for p in percentiles], index=percentile_labels))
                                                                                                                                                                                                                                                                                        .reset_index()
                                                                                                                                                                                                                                                                                        )

                                                                                                                                                                                                                                                                                        fig_p = go.Figure()
                                                                                                                                                                                                                                                                                        for label, color in zip(percentile_labels, colors):
                                                                                                                                                                                                                                                                                            fig_p.add_trace(go.Scatter(
                                                                                                                                                                                                                                                                                                    x=df_percentiles['A침oMes'],
                                                                                                                                                                                                                                                                                                            y=df_percentiles[label],
                                                                                                                                                                                                                                                                                                                    mode='lines+markers',
                                                                                                                                                                                                                                                                                                                            name=label,
                                                                                                                                                                                                                                                                                                                                    line=dict(color=color, width=3),
                                                                                                                                                                                                                                                                                                                                            marker=dict(size=6)
                                                                                                                                                                                                                                                                                                                                                ))

                                                                                                                                                                                                                                                                                                                                                fig_p.update_layout(
                                                                                                                                                                                                                                                                                                                                                    title="Curvas de Percentiles Mensuales: D칤as h치biles entre EXW Real y Fecha entrada Aduana",
                                                                                                                                                                                                                                                                                                                                                        xaxis_title="Mes/A침o",
                                                                                                                                                                                                                                                                                                                                                            yaxis_title="D칤as h치biles",
                                                                                                                                                                                                                                                                                                                                                                legend_title="Percentil",
                                                                                                                                                                                                                                                                                                                                                                    plot_bgcolor="white",
                                                                                                                                                                                                                                                                                                                                                                        font=dict(size=15),
                                                                                                                                                                                                                                                                                                                                                                            height=480,
                                                                                                                                                                                                                                                                                                                                                                                width=900
                                                                                                                                                                                                                                                                                                                                                                                )
                                                                                                                                                                                                                                                                                                                                                                                st.plotly_chart(fig_p)
```
