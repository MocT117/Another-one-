```python
import streamlit as st
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

st.set_page_config(page_title="Diferencia Fecha SO KMEX vs Fecha de PO", page_icon=":bar_chart:", layout="centered")

st.header("Análisis: Fecha SO KMEX vs Fecha de PO (días naturales)")
st.write("Sube tu archivo Excel")

archivo = st.file_uploader("Sube tu archivo Excel", type=["xlsx"])

if archivo:
    df = pd.read_excel(archivo)

        # Convertir columnas a fecha
            df['Fecha SO KMEX'] = pd.to_datetime(df['Fecha SO KMEX'], errors='coerce')
                df['Fecha de PO'] = pd.to_datetime(df['Fecha de PO'], errors='coerce')

                    # Filtrar filas válidas
                        df_so_po = df[df['Fecha SO KMEX'].notna() & df['Fecha de PO'].notna()].copy()

                            # Calcular diferencia en días naturales (Fecha de PO - Fecha SO KMEX)
                                df_so_po['Diferencia días naturales'] = (df_so_po['Fecha de PO'] - df_so_po['Fecha SO KMEX']).dt.days

                                    # === EXCLUIR VALORES NEGATIVOS ===
                                        df_so_po = df_so_po[df_so_po['Diferencia días naturales'] >= 0].copy()

                                            # Semáforo: On time <= 3 días naturales
                                                def status_so_po(dias):
                                                        if pd.isna(dias):
                                                                    return 'Sin fecha'
                                                                            elif dias <= 3:
                                                                                        return 'On time'
                                                                                                elif dias > 3:
                                                                                                            return 'Delay'
                                                                                                                    else:
                                                                                                                                return 'Revisar'
                                                                                                                                    df_so_po['Semáforo SO_PO'] = df_so_po['Diferencia días naturales'].apply(status_so_po)

                                                                                                                                        num_so_po = len(df_so_po)

                                                                                                                                            # Mostrar tabla SO vs PO
                                                                                                                                                st.subheader("Tabla (Fecha SO KMEX vs Fecha de PO)")
                                                                                                                                                    st.dataframe(df_so_po[['Fecha SO KMEX', 'Fecha de PO', 'Diferencia días naturales', 'Semáforo SO_PO']])

                                                                                                                                                        # Percentiles como texto SO_PO (ya SIN negativos)
                                                                                                                                                            serie_so_po = df_so_po['Diferencia días naturales'].dropna()
                                                                                                                                                                percentil_50_so_po = serie_so_po.quantile(0.5)
                                                                                                                                                                    percentil_75_so_po = serie_so_po.quantile(0.75)
                                                                                                                                                                        percentil_90_so_po = serie_so_po.quantile(0.9)
                                                                                                                                                                            percentil_95_so_po = serie_so_po.quantile(0.95)

                                                                                                                                                                                # Gauge chart AJUSTADO AL P95
                                                                                                                                                                                    st.subheader("Diferencia Fecha de PO vs Fecha SO KMEX (días naturales)")

                                                                                                                                                                                        if not serie_so_po.empty:
                                                                                                                                                                                                promedio_so_po = serie_so_po.mean()
                                                                                                                                                                                                    else:
                                                                                                                                                                                                            promedio_so_po = 0

                                                                                                                                                                                                                min_so_po = 0
                                                                                                                                                                                                                    max_so_po = int(percentil_95_so_po) + 2  # margen visual
                                                                                                                                                                                                                        tick_vals_so_po = list(range(min_so_po, max_so_po + 1, 5))

                                                                                                                                                                                                                            fig_so_po = go.Figure(go.Indicator(
                                                                                                                                                                                                                                    mode = "gauge+number",
                                                                                                                                                                                                                                            value = promedio_so_po,
                                                                                                                                                                                                                                                    number = {'suffix': " días", 'font': {'size': 36}},
                                                                                                                                                                                                                                                            title = {'text': "Fecha de PO vs Fecha SO KMEX"},
                                                                                                                                                                                                                                                                    gauge = {
                                                                                                                                                                                                                                                                                'axis': {'range': [min_so_po, max_so_po], 'tickmode': 'array', 'tickvals': tick_vals_so_po, 'tickfont': {'size': 16}},
                                                                                                                                                                                                                                                                                            'steps': [
                                                                                                                                                                                                                                                                                                            {'range': [min_so_po, percentil_50_so_po], 'color': "#ABEBC6"},   # Verde: Hasta mediana
                                                                                                                                                                                                                                                                                                                            {'range': [percentil_50_so_po, percentil_75_so_po], 'color': "#F9E79F"},   # Amarillo: Mediana a P75
                                                                                                                                                                                                                                                                                                                                            {'range': [percentil_75_so_po, percentil_95_so_po], 'color': "#F5B7B1"},   # Rojo: P75 a P95
                                                                                                                                                                                                                                                                                                                                                        ],
                                                                                                                                                                                                                                                                                                                                                                    'bar': {'color': "black", 'thickness': 0.25}
                                                                                                                                                                                                                                                                                                                                                                            }
                                                                                                                                                                                                                                                                                                                                                                                ))

                                                                                                                                                                                                                                                                                                                                                                                    st.plotly_chart(fig_so_po)
                                                                                                                                                                                                                                                                                                                                                                                        st.markdown(
                                                                                                                                                                                                                                                                                                                                                                                                f"<div style='text-align:center; font-size:18px; margin-top:-30px; margin-bottom:10px;'><b>Promedio considerando {num_so_po} diferencias</b></div>",
                                                                                                                                                                                                                                                                                                                                                                                                        unsafe_allow_html=True
                                                                                                                                                                                                                                                                                                                                                                                                            )
                                                                                                                                                                                                                                                                                                                                                                                                                st.info(
                                                                                                                                                                                                                                                                                                                                                                                                                        f"El gauge está ajustado al percentil 95 para evitar el sesgo de valores extremos. Los valores por arriba de {max_so_po} días existen pero no están representados aquí. Las diferencias negativas han sido excluidas del análisis."
                                                                                                                                                                                                                                                                                                                                                                                                                            )

                                                                                                                                                                                                                                                                                                                                                                                                                                # Resumen de percentiles
                                                                                                                                                                                                                                                                                                                                                                                                                                    st.subheader("Resumen de percentiles SO vs PO")
                                                                                                                                                                                                                                                                                                                                                                                                                                        st.write(f"Percentil 50 (Mediana): **{percentil_50_so_po:.2f}** días")
                                                                                                                                                                                                                                                                                                                                                                                                                                            st.write(f"Percentil 75: **{percentil_75_so_po:.2f}** días")
                                                                                                                                                                                                                                                                                                                                                                                                                                                st.write(f"Percentil 90: **{percentil_90_so_po:.2f}** días")
                                                                                                                                                                                                                                                                                                                                                                                                                                                    st.write(f"Percentil 95: **{percentil_95_so_po:.2f}** días")

                                                                                                                                                                                                                                                                                                                                                                                                                                                        # Histograma SO_PO
                                                                                                                                                                                                                                                                                                                                                                                                                                                            st.subheader("Histograma de diferencia de días naturales PO vs SO KMEX")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                fig_hist_so_po = px.histogram(df_so_po, x='Diferencia días naturales',
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                  nbins=10,
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    title="Histograma (frecuencia por valor)")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        st.plotly_chart(fig_hist_so_po)

                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            # Tendencia temporal (promedio por mes)
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                st.subheader("Tendencia histórica PO vs SO KMEX (promedio de días naturales por mes)")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    df_so_po['AñoMes'] = df_so_po['Fecha de PO'].dt.to_period('M').astype(str)
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        trend_so_po = df_so_po.groupby('AñoMes')['Diferencia días naturales'].mean().reset_index()
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            fig_trend_so_po = px.line(trend_so_po, x='AñoMes', y='Diferencia días naturales', markers=True,
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          title="Promedio de días naturales por mes")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                              st.plotly_chart(fig_trend_so_po)

                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                  st.info("Dashboard generado por Kai 🤖")

                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                  else:
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      st.warning("Sube un archivo para ver el dashboard.")
```
